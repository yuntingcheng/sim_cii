# make catalog from original one from SIDES website
http://cesam.lam.fr/sides/index/download

```python
from cosmo_tools import *
from sides.PYSIDES.pysides.gen_mass import *
from sides.PYSIDES.pysides.gen_sfr_props import *
from sides.PYSIDES.pysides.gen_magnification import *
from sides.PYSIDES.pysides.gen_fluxes import *
from sides.PYSIDES.pysides.gen_lines import *
```

```python
fname = '/Volumes/Drive/sim_cii/sides/PYSIDES/SIDES_Bethermin2017.csv'
df = pd.read_csv(fname, usecols = (1,2,3,4,5), header = None)
df.columns = ['redshift', 'ra', 'dec', 'Mhalo', 'Mstar']
```

```python
df = gen_sfr_props(df)
```

```python
df = gen_magnification(df)
```

```python
df = gen_fluxes(df)
```

```python
df = gen_lines(df)
```

```python
savename = '/Volumes/Drive/sim_cii/data_catalog/SIDES_cats'
pickle.dump(df, open(savename+'.p', "wb"))
```
